````python
import sqlite3
````

````python
def printSchema(connection):
    ### Source: http://stackoverflow.com/a/35092773/4765776
    for (tableName,) in connection.execute(
        """
        select NAME from SQLITE_MASTER where TYPE='table' order by NAME;
        """
    ):
        print("{}:".format(tableName))
        for (
            columnID, columnName, columnType,
            columnNotNull, columnDefault, columnPK,
        ) in connection.execute("pragma table_info('{}');".format(tableName)):
            print("  {id}: {name}({type}){null}{default}{pk}".format(
                id=columnID,
                name=columnName,
                type=columnType,
                null=" not null" if columnNotNull else "",
                default=" [{}]".format(columnDefault) if columnDefault else "",
                pk=" *{}".format(columnPK) if columnPK else "",
            ))
````

````python
conn = sqlite3.connect('wine.db')
c = conn.cursor()
print("Database schema:")
printSchema(conn)
````

Recreate the Normalized Tables from **Master1** and **Master2** as you did in the TP2

````python
%%sql
DROP TABLE IF EXISTS Wine;
CREATE TABLE Wine(
   NV INTEGER NOT NULL PRIMARY KEY,
   CRU varchar(255),
   DEGRE FLOAT,
   MILL INTEGER
);

DROP TABLE IF EXISTS Producer;
CREATE TABLE Producer(
   NP INTEGER NOT NULL PRIMARY KEY,
   NOM varchar(255),
   PRENOM varchar(255),
   REGION varchar(255)
);

DROP TABLE IF EXISTS Production;
CREATE TABLE Production(
   NV INTEGER NOT NULL,
   NP INTEGER NOT NULL,
   QTE INTEGER,
   PRIMARY KEY (NV,NP),
   FOREIGN KEY (NV) REFERENCES Wine(NV),
   FOREIGN KEY (NP) REFERENCES Producer(NB)
);

INSERT INTO Wine SELECT DISTINCT NV, CRU, DEGRE, MILL FROM Master1 WHERE NV IS NOT NULL;

INSERT INTO Producer SELECT DISTINCT NP, NOM, PRENOM, REGION FROM Master1 WHERE NP IS NOT NULL;

INSERT INTO Production SELECT DISTINCT NV, NP, QTE FROM Master1 WHERE (NP IS NOT NULL) AND (NV IS NOT NULL);

%%sql

DROP TABLE IF EXISTS Consumer;
CREATE TABLE Consumer(
   NB INTEGER NOT NULL PRIMARY KEY,
   NOM varchar(255),
   PRENOM varchar(255),
   TYPE varchar(255)
);

DROP TABLE IF EXISTS Consumption;
CREATE TABLE Consumption(
   NV INTEGER NOT NULL,
   NB INTEGER NOT NULL,
   DATES DATE NOT NULL,
   LIEU varchar(255) NOT NULL,
   QTE INTEGER,
   REGION varchar(255),
   PRIMARY KEY (NB, NV, DATES, LIEU)
   FOREIGN KEY (NV) REFERENCES Wine(NV),
   FOREIGN KEY (NB) REFERENCES Consumer(NB)
);


INSERT INTO Consumer SELECT DISTINCT NB, NOM, PRENOM, TYPE FROM Master2 WHERE NB IS NOT NULL;

INSERT INTO Consumption SELECT DISTINCT NV, NB, DATES, LIEU, QTE, REGION FROM Master2 WHERE (NV IS NOT NULL) AND (NB IS NOT NULL) AND (DATES IS NOT NULL) AND (LIEU IS NOT NULL);
````

# PART I: CREATE AND USE VIEWS

A view is a virtual table based on the result-set of an SQL statement. Views are stored in the database with an associated name.

Views are created following the syntax:

```mysql
CREATE VIEW view_name AS
SELECT column1, column2.....
FROM table_name
WHERE [condition];
```

An useful command:

```mysql
DROP VIEW IF EXISTS view_name;
```


__Warning:__ Use `DROP` with caution (only drop something if you are sure)

__Note:__ You will find some cells marked as "Test" that will help you check your work. Do NOT modify them. 

#### Exercise 1.1

Create a view "**bons_buveurs**" with the clients (buveurs) of type 'gros' or 'moyen'.

````python
%%sql
drop view if exists bons_buveurs;
create view bons_buveurs as 
select NB, NOM, PRENOM, TYPE
from consumer
where TYPE = 'moyen' or TYPE = 'gros'
````

````python
# Test
%sql SELECT * FROM bons_buveurs ORDER BY nb;
````

#### Exercise 1.2

Create the view "**buveurs_asec**" with clients (buveurs) who have not bought any wine.

````python
%%sql
drop view if exists buveurs_asec;
create view buveurs_asec as 
select consumer.NB, NOM, PRENOM, TYPE
from consumer left join consumption on (consumer.nb = consumption.nb)
where qte is null
````

````python
# Test
%sql SELECT * FROM buveurs_asec ORDER BY nb;
````

#### Exercise 1.3

Create the view "**buveurs_achats**" complementary to the previous one.

````python
%%sql
drop view if exists buveurs_achats;
create view buveurs_achats as 
select consumer.NB, NOM, PRENOM, TYPE
from consumer left join consumption on (consumer.nb = consumption.nb)
where qte is not null
group by nom
````

````python
# Test
%sql SELECT * FROM buveurs_achats ORDER BY nb;
````

#### Exercise 1.4

Create the view "**q83pl**" (LIEU, CRU, QTE_BUE) that provides by LIEU and CRU the total quantities bought in 1983 by all the clients (buveurs).

````python
%%sql
drop view if exists q83pl;
create view q83pl as 
select Lieu, Cru, sum(qte) as QTE_BUE
from wine join consumption on (wine.nv = consumption.nv)
where dates like '1983%'
group by lieu, cru
````

````python
# Test
%sql SELECT * FROM q83pl;
````

# PART II: UPDATE DATABASE CONTENT

The syntax for the `UPDATE` operation is:

```sql
UPDATE table_name
SET column1 = value1, column2 = value2...., columnN = valueN
WHERE [condition];
```

The syntax for the `INSERT` operation is:

```sql
INSERT INTO table_name [(column1, column2, column3,...columnN)]  
VALUES (value1, value2, value3,...valueN);
```

Database updates are commited automatically in Jupyter/Python. _Transactions_ are an option to control and reverse changes. Additionally we can just reload a backup of the database (NOT an option in deployed systems)

__Note:__ Different to other Database Management Systems, SQLite views are read-only and so you can not execute a `DELETE`, `INSERT` or `UPDATE` statement on a view.

#### Exercise 2.2

Update the table you used to create "**bons_buveurs**": Change the "type" to 'gros' if the total of quantities bought is over 100.

Find the instances to update (schema may differ from the one in your table)

````python
%%sql
select consumption.NB as NB, NOM, PRENOM, TYPE, sum(qte) as TOTAL
from consumer join consumption on (consumer.nb = consumption.nb)
where type != 'gros'
group by consumption.nb
having sum(qte) > 100
order by consumption.nb
````

Update instances

````python
%%sql
UPDATE consumer
SET TYPE = 'gros'
where nb in (select consumption.nb 
            from  consumer join consumption on (consumer.nb = consumption.nb)
            where type != 'gros' 
            group by consumption.nb
            having sum(qte) > 100
            order by consumption.nb )
````
